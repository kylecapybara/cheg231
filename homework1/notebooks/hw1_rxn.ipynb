First I'll copy over the original plot for reference.

```python
import numpy as np
import matplotlib.pyplot as plt

tmax = 10 # Total length (in time) of the simulation

### Initial Conditions  ###

CA0 = 1.  # mol/L
CB0 = 0.1 # mol/L 

k1 = 0.5 # Rate constants (units of inverse time)
k2 = 0.11 
```

```python
t= np.linspace(0,tmax,100)
CA = CA0 * np.exp(-k1 * t)
CB = k1*CA0 / (k2 - k1) * (np.exp(-k1*t) - np.exp(-k2*t)) + CB0 * np.exp(-k2*t)
CC = CA0 + CB0 - CA - CB
```

```python
fig, ax = plt.subplots()
plt.title('Concentration of species A, B and C as a function of time')
ax.plot(t,CA, label='C$_A$')
ax.plot(t,CB, label='C$_B$')
ax.plot(t,CC, label='C$_C$')
ax.set(xlabel='Time',ylabel='Concentrations C$_i$/C$_A0$')
ax.legend(loc='best')
ax.set_xlim(t[0],t[-1])
ax.set_ylim(0,1)
plt.show()
```
