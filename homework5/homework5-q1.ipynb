The molar entropy can be calculated using this equation
$$
    \underbar{S} = Cp^* \ln (T)
$$

Where $Cp^*$ is the molar heat capacity, and $T$ is the temperature.

```python
import numpy as np
import matplotlib.pyplot as plt 

def co2_cp(T):
    return 22.243 + (5.977e-2 * T) - (3.499e-5 * T**2) + (7.464e-9 * T**3)

x = np.arange(273, 1800)
y = co2_cp(x) * np.log(x)
y_ideal = 37.15 * np.log(x)

fig,ax = plt.subplots(figsize=(8,6), dpi=100)

ax.plot(x,y, label='Using Appendix II')
ax.plot(x,y_ideal, label='Using predicted Cp')
ax.set_xlabel('Temperature (K)')
ax.set_ylabel('Molar Entropy (J/mol$\cdot$K)')
ax.set_xlim(273,1800)
ax.set_title('CO$_2$ Molar Entropy (J/mol$\cdot$K) vs. Temperature (K)');
```

Clearly the constant value of Cp predicts fairly different molar entropy values, especially as the temperature increases.
