# honors section boltzmann distribution

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from scipy.constants import k
import matplotlib

def boltzmannV(T, m):
    v = np.linspace(-2000, 2000, 500)
    f_v = (m / (2 * np.pi * k * T))**0.5 * np.exp(-m * v**2 / (2 * k * T))
    return v, f_v

temps = np.arange(20, 300, 20)  
m = 2.3258671e-26

plt.figure(figsize=(10, 6))

colormap = matplotlib.colormaps.get_cmap('coolwarm')

for i,T in enumerate(temps):
    v, f_v = boltzmannV(T, m)
    plt.plot(v, f_v, label=f'T={T}K', color=colormap(i / 10.0))

plt.xlabel('Velocity (m/s)')
plt.ylabel('Probability Density')
plt.title('Maxwell-Boltzmann Velocity Distribution for Different Temperatures')
plt.legend()
plt.show()
```

```python
def boltzmannS(T, m):
    v = np.linspace(0, 2000, 500)
    f_v = (m / (2 * np.pi * k * T))**1.5 * 4 * np.pi * v**2 * np.exp(-m * v**2 / (2 * k * T))
    return v, f_v

plt.figure(figsize=(10, 6))

for i,T in enumerate(temps):
    v, f_v = boltzmannS(T, m)
    plt.plot(v, f_v, label=f'T={T}K', color=colormap((i)/10))

plt.xlabel('Speed (m/s)')
plt.ylabel('Probability Density')
plt.title('Maxwell-Boltzmann Speed Distribution for Different Temperatures')
plt.legend()
plt.show()
```
